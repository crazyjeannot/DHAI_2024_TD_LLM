## CONVERT TO HTML

```python
path_jean_html = "TO_ANNOTATE/HTML/JEAN_HTML/"
path_pedro_html = "TO_ANNOTATE/HTML/PEDRO_HTML/"
path_stage1_html = "TO_ANNOTATE/HTML/STAGE1_HTML/"
path_stage2_html = "TO_ANNOTATE/HTML/STAGE2_HTML/"
```

```python
def convert_repository_to_html(input_directory, output_directory, line_break_interval=10):
    try:
        # Create the output directory if it doesn't exist
        if not os.path.exists(output_directory):
            os.makedirs(output_directory)

        # Iterate through all files in the input directory
        for filename in os.listdir(input_directory):
            if filename.endswith(".txt"):
                # Read the content from each text file
                input_file_path = os.path.join(input_directory, filename)
                with open(input_file_path, 'r') as txt_file:
                    txt_content = txt_file.read()

                # Apply basic styling and insert line breaks
                styled_content = "\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n"
                tokens_processed = 0
                for token in txt_content.split():
                    styled_content += f"{token} "
                    tokens_processed += 1
                    if tokens_processed % line_break_interval == 0:
                        styled_content += '\n'

                # Create an HTML file with the styled content
                # Create an HTML file with the styled content
                output_file_path = os.path.join(output_directory, f"{os.path.splitext(filename)[0]}.html")
                html_content = f"""
                <!DOCTYPE html>
                <html>
                <head>
                    <title>{filename} - Text to HTML Conversion</title>
                    <style>
                        body {{
                            font-size: 20px;
                            line-height: 1.5;
                            display: flex;
                            align-items: center;
                            justify-content: center;
                            height: 100vh;
                            margin: 0;
                        }}
                        pre {{
                            white-space: pre-wrap;
                            font-family: 'Courier New', Courier, monospace;
                            text-align: left;
                        }}
                    </style>
                </head>
                <body>
                    <pre>{styled_content}</pre>
                </body>
                </html>
                """


                # Write the HTML content to the output file
                with open(output_file_path, 'w') as html_file:
                    html_file.write(html_content)

                #print(f"Conversion successful. HTML file saved as {output_file_path}")

    except Exception as e:
        print(f"Error: {e}")
```

# Chunkise Texts

```python
def chunkize_text_files(N, M, path_to_folder, new_dir):
    # Create a new directory if it doesn't exist
    os.makedirs(new_dir, exist_ok=True)

    # Iterate through each text file in the given folder
    for filename in tqdm(os.listdir(path_to_folder), total=len(os.listdir(path_to_folder))):
        if filename.endswith(".txt"):
            file_path = os.path.join(path_to_folder, filename)
            with open(file_path, 'r', encoding='utf-8') as file:
                content = file.read()
                tokens = content.split()

                # Create overlapping chunks
                for i in range(0, len(tokens) - N, M):
                    chunk = tokens[i:i + N]

                    # Save the chunk to a new text file
                    chunk_filename = f"{filename}_chunk_{i // M}.txt"
                    chunk_filepath = os.path.join(new_dir, chunk_filename)
                    with open(chunk_filepath, 'w', encoding='utf-8') as chunk_file:
                        chunk_file.write(' '.join(chunk))
                        
```

```python
def random_copy_files(source_folder, destination_folder, num_files=10000):
    # Get a list of all files in the source folder
    all_files = os.listdir(source_folder)

    # Ensure that the number of files to copy is not greater than the total number of files in the source folder
    num_files = min(num_files, len(all_files))

    # Randomly select num_files files
    selected_files = random.sample(all_files, num_files)

    # Create the destination folder if it doesn't exist
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    # Copy selected files to the destination folder
    for file_name in selected_files:
        source_path = os.path.join(source_folder, file_name)
        destination_path = os.path.join(destination_folder, file_name)
        shutil.copy2(source_path, destination_path)
```

```python
def organize_by_subgenre(file_names, subgenres, path_folder):
    # Create subgenre folders
    for subgenre in subgenres:
        subgenre_folder = path.join(path_folder, subgenre)
        os.makedirs(subgenre_folder, exist_ok=True)

    # Iterate through files in the main folder
    for file_name in os.listdir(path_folder):
        file_path = path.join(path_folder, file_name)
        file_name = path.splitext(path.basename(file_name))[0]
        # Check if the item is a file
        if path.isfile(file_path):
            # Check if the file_name is in the provided list
            if file_name in file_names:
                # Get the index of the file_name in the list
                index = file_names.index(file_name)

                # Get the corresponding subgenre
                file_subgenre = subgenres[index]

                # Create destination folder for the subgenre if not exists
                subgenre_folder = path.join(path_folder, file_subgenre)
                dest_folder = path.join(subgenre_folder, file_name)

                # Copy the file to the subgenre folder
                shutil.copy2(file_path, dest_folder+".txt")
                print(f"File {file_name} copied to {file_subgenre} folder.")
```

## PREP FINE TUNING DATASET

```python
import os
import random
import shutil
```

```python
def merge_files(input_folder, output_file, max_size):
    file_list = os.listdir(input_folder)
    random.shuffle(file_list)

    merged_size = 0
    with open(output_file, 'w') as merged_file:
        for file_name in file_list:
            file_path = os.path.join(input_folder, file_name)
            file_size = os.path.getsize(file_path)

            if merged_size + file_size <= max_size:
                with open(file_path, 'r') as current_file:
                    merged_file.write(current_file.read())
                merged_size += file_size
            else:
                break
```
