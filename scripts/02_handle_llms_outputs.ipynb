```python
import pandas as pd
import numpy as np
import re
from sklearn.utils import shuffle
```

```python
MIXTRAL_ANNOT = pd.read_csv("/home/crazyjeannot/Documents/cours/DHAI_intensive_week_llms/DHAI_2024_TD_LLM/data/MIXTRAL_ANNOT_DHAI_MAIN.csv")
```

```python
def categorize_labels(df):
    categories = []
    for label in df['label']:
        label_lower = label.lower()
        if 'exploration' in label_lower or 'exp' in label_lower:
            categories.append('EXPLORATION')
            pass
        elif 'danger' in label_lower or 'dan' in label_lower:
            categories.append('DANGER')
            pass
        else:
            categories.append('OTHER')

    df['label_net'] = categories

    return df.drop(["label"], axis=1)
```

```python
MIXTRAL_LABELS = categorize_labels(MIXTRAL_ANNOT)
```

```python
MIXTRAL_LABELS.label_net.value_counts()
```

```python
MIXTRAL_LABELS_SORTED = MIXTRAL_LABELS.sort_values(by='doc')
```

```python
MIXTRAL_LABELS_SORTED.reset_index(drop=True, inplace=True)
```
